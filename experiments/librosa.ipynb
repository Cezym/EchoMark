```python
import librosa
import matplotlib.pyplot as plt
import numpy as np

# Wczytaj przykładowy plik audio
y, sr = librosa.load(librosa.ex('trumpet'))

# Oblicz wektor czasu
t = np.arange(0, len(y)) / sr

# Rysuj wykres
plt.figure(figsize=(10, 4))
plt.plot(t, y)
plt.xlabel('Czas (s)')
plt.ylabel('Amplituda')
plt.title('Sygnał audio (trumpet)')
plt.grid(True)
plt.show()
```

```python
import librosa
import sounddevice as sd
import matplotlib.pyplot as plt
import numpy as np

example_name = 'nutcracker'

# Wczytaj audio
audio, sr = librosa.load(librosa.ex(example_name))

# Oblicz FFT (widmo częstotliwości)
fft = np.fft.fft(audio)
magnitude = np.abs(fft)
frequencies = np.fft.fftfreq(len(fft), 1/sr)

# Wyświetl dominującą częstotliwość
max_freq = frequencies[np.argmax(magnitude)]
print(f"Dominująca częstotliwość: {max_freq:.2f} Hz")
max_posible_freq = sr/2
print(f"Maksymalna możliwa częstotliwość wg. tw. Nyquista-Shannona {max_posible_freq} Hz dla sample rate {sr} Hz")
t = np.arange(0, len(audio)) / sr

# Rysuj wykres
plt.figure(figsize=(10, 4))
plt.plot(t, audio)
plt.xlabel('Czas (s)')
plt.ylabel('Amplituda')
plt.title(f'Sygnał audio ({example_name})')
plt.grid(True)
plt.show()
```

```python
import librosa
import sounddevice as sd
import numpy as np

# Parametry
sr = 44100  # Częstotliwość próbkowania
duration = 1  # Czas trwania w sekundach
freq = 440  # Częstotliwość tonu (A4)

# Generuj ton
t = np.linspace(0, duration, int(sr * duration))
audio = 0.5 * np.sin(2 * np.pi * freq * t)

# Odtwórz ton
sd.play(audio, sr)
sd.wait()

# Rysuj wykres
plt.figure(figsize=(10, 4))
plt.plot(t, audio)
plt.xlabel('Czas (s)')
plt.ylabel('Amplituda')
plt.title('Sygnał audio (A4)')
plt.grid(True)
plt.xlim(0, 0.004)
plt.show()
```
